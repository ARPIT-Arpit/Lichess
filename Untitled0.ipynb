```python
import pandas as pd
import networkx as nx

def preprocess_csv(file_path):
    """
    Preprocess a CSV file containing URLs and labels.

    Parameters:
    - file_path: Path to the CSV file

    Returns:
    - graph: Directed graph (NetworkX DiGraph)
    - seed_nodes: List of legitimate websites as seed nodes
    """
    # Load the dataset
    df = pd.read_csv(file_path)

    # Check the structure of the dataset
    print("Dataset Head:")
    print(df.head())

    # Validate required columns
    if 'URL' not in df.columns or 'Label' not in df.columns:
        raise ValueError("The dataset must have 'URL' and 'Label' columns.")

    # Split URLs into legitimate and phishing
    legitimate_sites = df[df['Label'] == 0]['URL'].tolist()
    phishing_sites = df[df['Label'] == 1]['URL'].tolist()

    # Create a graph with nodes and hypothetical links (edges)
    # Here, legitimate sites are connected randomly for demonstration
    graph = nx.DiGraph()
    graph.add_nodes_from(legitimate_sites + phishing_sites)

    # Add some sample edges (you can customize this part based on your dataset)
    for site in legitimate_sites:
        for other_site in legitimate_sites:
            if site != other_site:  # Example: Legitimate sites link to each other
                graph.add_edge(site, other_site)

    for site in phishing_sites:
        for legit_site in legitimate_sites:
            graph.add_edge(site, legit_site)  # Example: Phishing sites link to legitimate sites

    # Legitimate sites are used as seed nodes
    seed_nodes = legitimate_sites

    print(f"Graph created with {graph.number_of_nodes()} nodes and {graph.number_of_edges()} edges.")
    return graph, seed_nodes

# Example usage
file_path = "phishing_dataset.csv"  # Replace with your CSV file path
graph, seed_nodes = preprocess_csv(file_path)

# Use TrustRank
scores = trustrank(graph, seed_nodes)
print("TrustRank Scores:")
print(scores)
```

```python
import pandas as pd
import networkx as nx
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Load datasets
benign_file = "1.Benign_list_big_final.csv"
legitimate_file = "3.legitimate.csv"
phishing_file = "4.phishing.csv"

# Load data
benign_df = pd.read_csv(benign_file)
legitimate_df = pd.read_csv(legitimate_file)
phishing_df = pd.read_csv(phishing_file)

# Ensure column consistency
benign_links = benign_df['URL'].tolist()
legitimate_links = legitimate_df['URL'].tolist()
phishing_links = phishing_df['URL'].tolist()

# Create graph
def create_graph(legitimate_links, phishing_links):
    graph = nx.DiGraph()

    # Add nodes
    graph.add_nodes_from(legitimate_links + phishing_links)

    # Add edges (hypothetical relationships for demonstration)
    # Legitimate sites linking to each other
    for site in legitimate_links:
        for other_site in legitimate_links:
            if site != other_site:
                graph.add_edge(site, other_site)

    # Phishing sites linking to legitimate sites
    for phish_site in phishing_links:
        for legit_site in legitimate_links:
            graph.add_edge(phish_site, legit_site)

    return graph

graph = create_graph(legitimate_links, phishing_links)

# Define TrustRank algorithm
def trustrank(graph, seed_nodes, alpha=0.85, max_iter=100, tol=1e-6):
    """
    Compute TrustRank scores for a given web graph.
    """
    n = len(graph.nodes)
    scores = np.zeros(n)
    seed_set = set(seed_nodes)

    # Initialize scores (seed nodes get equal weight, others get zero)
    node_list = list(graph.nodes)
    for i, node in enumerate(node_list):
        if node in seed_set:
            scores[i] = 1 / len(seed_set)

    scores = np.array(scores)
    prev_scores = np.zeros_like(scores)
    adjacency_matrix = nx.to_numpy_array(graph)
    out_degree = adjacency_matrix.sum(axis=1)
    teleport = scores.copy()

    # Normalize rows of the adjacency matrix
    for i in range(len(out_degree)):
        if out_degree[i] > 0:
            adjacency_matrix[i] /= out_degree[i]

    for _ in range(max_iter):
        # TrustRank update rule
        scores = alpha * adjacency_matrix.T @ scores + (1 - alpha) * teleport

        # Convergence check
        if np.linalg.norm(scores - prev_scores) < tol:
            break
        prev_scores = scores.copy()

    # Normalize scores to sum to 1
    scores /= scores.sum()

    # Map scores to nodes
    trustrank_scores = {node_list[i]: scores[i] for i in range(n)}
    return trustrank_scores

# Use legitimate links as seed nodes
seed_nodes = legitimate_links

# Compute TrustRank scores
scores = trustrank(graph, seed_nodes)

# Evaluate TrustRank results
def evaluate(scores, benign_links, phishing_links):
    """
    Evaluate TrustRank performance.
    """
    # Assign scores to the links
    benign_scores = {link: scores.get(link, 0) for link in benign_links}
    phishing_scores = {link: scores.get(link, 0) for link in phishing_links}

    # Prepare ground truth and predictions
    y_true = [0] * len(benign_links) + [1] * len(phishing_links)
    y_pred = [1 if benign_scores.get(link, 0) < 0.5 else 0 for link in benign_links] + \
             [1 if phishing_scores.get(link, 0) < 0.5 else 0 for link in phishing_links]

    # Metrics
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)

    print(f"Accuracy: {accuracy:.2f}")
    print(f"Precision: {precision:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"F1 Score: {f1:.2f}")

# Evaluate TrustRank
evaluate(scores, benign_links, phishing_links)
```

```python
import pandas as pd
import networkx as nx
import numpy as np
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# File paths
benign_file = "1.Benign_list_big_final.csv"
legitimate_file = "3.legitimate.csv"
phishing_file = "4.phishing.csv"

# Load datasets
benign_df = pd.read_csv(benign_file, header=None)  # No header in this file
legitimate_df = pd.read_csv(legitimate_file)
phishing_df = pd.read_csv(phishing_file)

# Extract URLs
benign_links = benign_df[0].tolist()  # URLs in the first column
legitimate_links = legitimate_df['Domain'].tolist()
phishing_links = phishing_df['Domain'].tolist()

# Print the number of links for verification
print(f"Benign Links: {len(benign_links)}")
print(f"Legitimate Links: {len(legitimate_links)}")
print(f"Phishing Links: {len(phishing_links)}")
```

```python
def create_graph(legitimate_links, phishing_links):
    """
    Create a directed graph for TrustRank.

    Parameters:
    - legitimate_links: List of legitimate website URLs.
    - phishing_links: List of phishing website URLs.

    Returns:
    - graph: A NetworkX DiGraph object.
    """
    graph = nx.DiGraph()

    # Add nodes
    graph.add_nodes_from(legitimate_links + phishing_links)

    # Add edges: Legitimate sites linking to each other
    for site in legitimate_links:
        for other_site in legitimate_links:
            if site != other_site:
                graph.add_edge(site, other_site)

    # Add edges: Phishing sites linking to legitimate sites
    for phish_site in phishing_links:
        for legit_site in legitimate_links:
            graph.add_edge(phish_site, legit_site)

    return graph

# Create the graph
graph = create_graph(legitimate_links, phishing_links)
print(f"Graph has {graph.number_of_nodes()} nodes and {graph.number_of_edges()} edges.")
```

```python
def trustrank(graph, seed_nodes, alpha=0.85, max_iter=100, tol=1e-6):
    """
    Compute TrustRank scores for a given web graph.

    Parameters:
    - graph: NetworkX graph.
    - seed_nodes: List of seed nodes (trusted websites).
    - alpha: Damping factor for teleport probability.
    - max_iter: Maximum number of iterations.
    - tol: Convergence tolerance.

    Returns:
    - scores: Dictionary of TrustRank scores for each node.
    """
    nodes = list(graph.nodes)
    n = len(nodes)
    node_index = {node: i for i, node in enumerate(nodes)}

    # Initialize scores
    scores = np.zeros(n)
    for seed in seed_nodes:
        if seed in node_index:
            scores[node_index[seed]] = 1 / len(seed_nodes)

    adjacency_matrix = nx.to_numpy_array(graph)
    out_degree = adjacency_matrix.sum(axis=1)

    # Normalize rows of adjacency matrix
    for i in range(len(out_degree)):
        if out_degree[i] > 0:
            adjacency_matrix[i] /= out_degree[i]

    prev_scores = np.zeros_like(scores)
    for _ in range(max_iter):
        # TrustRank update rule
        scores = alpha * adjacency_matrix.T @ scores + (1 - alpha) * scores
        if np.linalg.norm(scores - prev_scores) < tol:
            break
        prev_scores = scores.copy()

    # Normalize scores
    scores /= scores.sum()

    return {nodes[i]: scores[i] for i in range(n)}

# Use legitimate links as seed nodes
seed_nodes = legitimate_links
scores = trustrank(graph, seed_nodes)
```

```python
def evaluate(scores, benign_links, phishing_links):
    """
    Evaluate TrustRank performance.

    Parameters:
    - scores: Dictionary of TrustRank scores.
    - benign_links: List of benign URLs.
    - phishing_links: List of phishing URLs.

    Returns:
    - None (prints evaluation metrics).
    """
    # Assign scores to benign and phishing links
    benign_scores = {link: scores.get(link, 0) for link in benign_links}
    phishing_scores = {link: scores.get(link, 0) for link in phishing_links}

    # Prepare ground truth and predictions
    y_true = [0] * len(benign_links) + [1] * len(phishing_links)
    y_pred = [1 if benign_scores.get(link, 0) < 0.5 else 0 for link in benign_links] + \
             [1 if phishing_scores.get(link, 0) < 0.5 else 0 for link in phishing_links]

    # Metrics
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)

    print(f"Accuracy: {accuracy:.2f}")
    print(f"Precision: {precision:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"F1 Score: {f1:.2f}")

# Evaluate TrustRank
evaluate(scores, benign_links, phishing_links)
```
